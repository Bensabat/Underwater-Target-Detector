### Actual preprocessing

```python
def image_preprocessing(path):
  # Open image
  try:
    img = cv.imread(path, 0).astype(np.uint8)
  except:
    return None
  
  # Scaling 
  img2 = cv.resize(img, (WIDTH, HEIGHT), interpolation=cv.INTER_CUBIC)
  
  # CLAHE (Contrast Limited Adaptive Histogram Equalization)
  img2 = cv.createCLAHE(clipLimit=2.0, tileGridSize=(8,8)).apply(img2)

  # Median filter
  img2 = cv.medianBlur(img2, KERNEL_SIZE)

  # Sobel filtering: apply cv.Sobel to have output dtype = cv.CV_64F (=float64). 
  # Then take its absolute and convert to cv2.CV_8U (=uint8).
  #sobel64f = cv.Sobel(img2,cv.CV_64F,1,0,ksize=KERNEL_SIZE)
  #abs_sobel64f = np.absolute(sobel64f)
  #img2 = np.uint8(abs_sobel64f)
  
  return img2
```

## Data loading

```python
# Build the training dataset (samples & labels) from the 4 image subdirectories
samples = None
checkerboard_labels = None
target_labels = None
paths = ['../../data/snap/seabed/',
         '../../data/snap/checkerboard',
         '../../data/snap/target/', 
         '../../data/snap/checkerboard_and_target/']
for i in range(4):
  # checkerboard_label: i=1 and i=3 stand for samples with targets in them,
  # so we define the new label as a the NumPy array [[1]]. i=0 and i=2 yield
  # the new label [[0]].
  new_checkerboard_label = (i % 2) * np.ones((1,1), dtype=np.int32)
  
  # new_target_label: works similarly, except that i=0 and i=1 give [[0]],
  # and i=2 and i=3 give [[1]].
  new_target_label = (i // 2) * np.ones((1,1), dtype=np.int32)
  
  # Walk through all 4 directories
  for dirname, dirnames, filenames in os.walk(paths[i]):
    for filename in filenames:
      full_path = os.path.join(dirname, filename)
      
      # process the image given by full path 
      new_sample = image_preprocessing(full_path).astype(np.float32)
      new_sample.resize(1, HEIGHT * WIDTH)
      
      # do data augmentation on it
      # TODO
      
      # append new samples and labels as rows to the samples and labels arrays
      if samples is None:
        samples = new_sample
        checkerboard_labels = new_checkerboard_label
        target_labels = new_target_label 
      else:
        samples = np.append(samples, new_sample, axis=0)
        checkerboard_labels = np.append(checkerboard_labels,
                                        new_checkerboard_label, axis=0)
        target_labels = np.append(target_labels,
                                  new_target_label, axis=0)
```

## Checkerboard and target detection

### Training (with 2 SVMs)

```python
# SVM definition
def svm_create(P=0.2, C=2.67):
  svm = cv.ml.SVM_create()
  svm.setKernel(cv.ml.SVM_LINEAR)
  svm.setP(P)
  svm.setType(cv.ml.SVM_C_SVC)
  svm.setC(C)
  return svm
```

```python
# Training
checkerboard_svm = svm_create()
checkerboard_svm.train(samples, cv.ml.ROW_SAMPLE, checkerboard_labels)

target_svm = svm_create()
target_svm.train(samples, cv.ml.ROW_SAMPLE, target_labels)

pass
```

### Detection

```python
def has_feature(path, svm): # feature = checkerboard or target
  img = image_preprocessing(path).reshape(1, HEIGHT * WIDTH).astype(np.float32)
  response = svm.predict(img)[1][0, 0]
  return response == 1

def has_checkerboard(path, svm=checkerboard_svm):
  return has_feature(path, svm)

def has_target(path, svm=target_svm):
  return has_feature(path, svm)
```
